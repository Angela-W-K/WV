```python
# Convert float columns to string
float_columns = data.select_dtypes(include=['float64']).columns
data[float_columns] = data[float_columns].astype(str)
```

```python
# Use one-hot encoding for categorical variables
data = pd.get_dummies(data, columns=['SEX', 'RESIDENCE', 'DEWORMED', 'DISABILITY', 'BREAST FEEDING'], drop_first=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pie chart for 'STATUS'
status_counts = data['STATUS'].value_counts()
plt.figure(figsize=(6,6))
status_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title('Distribution of STATUS')
plt.ylabel('')
plt.show()


```

```python
# Line plot for 'AGE in months' vs 'STATUS'
sns.lineplot(data=data, x='AGE', y='STATUS')
plt.title('AGE vs STATUS')
plt.show()
```

```python
# Scatter plot of AGE and MUAC
sns.scatterplot(data=data, x='AGE', y='MUAC', hue='STATUS')
plt.title('AGE vs MUAC')
plt.show()
```

```python
# Histogram for 'AGE in months'
data['AGE'].hist(bins=20)
plt.title('AGE Distribution')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Define dependent (y) and independent variables (X)
X = data[['AGE', 'MUAC', 'SEX_male', 'DEWORMED_yes']]  # Example features
y = data['STATUS']  # Target variable

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Predict on the test set
y_pred = model.predict(X_test)

```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Classification Report (Precision, Recall, F1-Score)
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

```

```python
import seaborn as sns

# Plot confusion matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
import pickle

# Save the model to a file
with open('logistic_regression_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
# Load the saved model
with open('logistic_regression_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)

# Make predictions using the loaded model
new_predictions = loaded_model.predict(X_test)
print(new_predictions)
```
